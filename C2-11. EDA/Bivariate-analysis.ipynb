```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
auag = pd.read_csv('EDA_Gold_Silver_prices.csv')
auag['Month'] = pd.to_datetime(auag['Month'], format='%b-%y')
auag.head(2)
```

```python
sns.pairplot(auag[['SilverPrice', 'GoldPrice']])
plt.show()
```

```python
c = auag[['SilverPrice', 'GoldPrice']].corr()
c
```
