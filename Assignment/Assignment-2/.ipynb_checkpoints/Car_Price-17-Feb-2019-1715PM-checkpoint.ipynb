### PROJECT BRIEF###

A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts. 

 

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:

1. Which variables are significant in predicting the price of a car?
2. How well those variables describe the price of a car?

Based on various market surveys, the consulting firm has gathered a large dataset of different types of cars across the Americal market. 

### BUSINESS GOAL ### 

You are required to model the price of cars with the available independent variables. It will be used by the management to understand how exactly the prices vary with the independent variables. They can accordingly manipulate the design of the cars, the business strategy etc. to meet certain price levels. Further, the model will be a good way for management to understand the pricing dynamics of a new market.

```python
car_price.head()
```

### STEP 4 : SPLITTING THE DATA INTO TRAINING AND TESTING SETS

As we know that the first basic step for regression is performing a train-test split.

```python
np.random.seed(0)
car_price_train, car_price_test = train_test_split(car_price, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
car_price_train.info()
```

#### RESCALING THE FEATURES

As we know that scaling doesn't impact your model. Here we can see that except for peakrpm, all the columns have small integer values. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling
2. Standardisation (mean-0, sigma-1)

This time, we will use MinMax scaling.

```python
scaler = MinMaxScaler()

fields_to_tfm = ['symboling', 'price', 'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', \
                 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']

car_price_train[fields_to_tfm] = scaler.fit_transform(car_price_train[fields_to_tfm])
```

##### DIVIDING INTO X AND Y SETS FOR THE MODEL BUILDING

```python
y_train = car_price_train.pop('price') # Taking out the price data for y-axis
X_train = car_price_train
```

```python
X = X_train.drop('fuel_system_2bbl',axis = 1, inplace = False) # high p-value = 0.926 | low VIF = 4.23
X = X_train.drop('fuel_system_mpfi',axis = 1, inplace = False) # high p-value = 0.895 | low VIF = 4.63
X = X_train.drop('aspiration',axis = 1, inplace = False) # high p-value = 0.886 | low VIF = 3.40
X = X_train.drop('carheight',axis = 1, inplace = False) # high p-value = 0.884 | low VIF = 2.65
X = X_train.drop('compressionratio',axis = 1, inplace = False) # high p-value = 0.828 | high VIF = 81.39
X = X_train.drop('fueltype',axis = 1, inplace = False) # high p-value = 0.724 | high VIF = 97.50
X = X_train.drop('boreratio',axis = 1, inplace = False) # high p-value = 0.721 | low VIF = 4.00
X = X_train.drop('citympg',axis = 1, inplace = False) # high p-value = 0.677 | high VIF = 7.21
X = X_train.drop('curbweight',axis = 1, inplace = False) # high p-value = 0.536 | high VIF =27.57
X = X_train.drop('horsepower',axis = 1, inplace = False) # high p-value = 0.384 | high VIF = 12.98
X = X_train.drop('stroke',axis = 1, inplace = False) # high p-value = 0.362 | low VIF = 1.98
X = X_train.drop('OHC_type_engine',axis = 1, inplace = False) # high p-value = 0.306 | low VIF = 2.17
X = X_train.drop('symboling',axis = 1, inplace = False) # high p-value = 0.240 | low VIF = 3.03
X = X_train.drop('car_body',axis = 1, inplace = False) # high p-value = 0.211 | low VIF = 1.69
X = X_train.drop('wheelbase',axis = 1, inplace = False) # high p-value = 0.188 | high VIF = 10.52
X = X_train.drop('cylinder_number_less_than_four',axis = 1, inplace = False) # high p-value = 0.165 | high VIF = 5.07
X = X_train.drop('peakrpm',axis = 1, inplace = False) # high p-value = 0.118 | low VIF = 2.72
X = X_train.drop('enginesize',axis = 1, inplace = False) # low p-value = 0.007 | high VIF = 16.10
X = X_train.drop('carwidth',axis = 1, inplace = False) # low p-value = 0.007 | high VIF = 7.42
```

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train['enginesize'])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Check the parameters obtained

lr.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], -0.073 + 1.211*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Print a summary of the linear regression model obtained
print(lr.summary())
```

### Checking VIF

Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model. The formula for calculating `VIF` is:

### $ VIF_i = \frac{1}{1 - {R_i}^2} $

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### STEP 5 : BUILDING A LINEAR MODEL

Fit a regression line through the training data using `statsmodels`. Remember that in `statsmodels`, you need to explicitly fit a constant using `sm.add_constant(X)` because if we don't perform this step, `statsmodels` fits a regression line passing through the origin, by default.

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['horsepower','highwaympg']])

# Create a first fitted model
lr_1 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Check the parameters obtained

lr_1.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.105 + 0.864*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Print a summary of the linear regression model obtained
print(lr_1.summary())
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step 6: Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
X_train_lm.head()
```

```python
y_train_price = lr_1.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 30)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```
