```python
import numpy as np
import pandas as pd

companies_df = pd.read_csv("C:\\Amar\\Personal\\Studies\\PG\ML\\Assignment\\companies.txt", sep='\t', lineterminator='\n', encoding='unicode_escape')
rounds2_df = pd.read_csv("C:\\Amar\\Personal\\Studies\\PG\ML\\Assignment\\rounds2.csv", encoding='unicode_escape')

companies_df.head()
```

```python
rounds2_df.head()
```

```python
#list of unique companies in round2 and companies file
print("list of unique companies in round2: ")
print(len(rounds2_df['company_permalink'].unique().tolist()))


print("list of unique companies: ")
print(len(companies_df['permalink'].unique().tolist()))
```

```python
#Checking if there are records in rounds2 dataframe which are not present in companies dataframe.
rounds2_df[~rounds2_df.isin(companies_df)].dropna().count()
```

```python
#merging companies dataframe into round2 dataframe to find the number of observations in this merged dataframe

# before merging the values in both the dataframes are not in same case. Hence it is not creating a merged record.
# to mitigate this am creating a new column for permalink and company_permalink in both the dataframes, whose 
# values will be in lowercase.

companies_df['permalink_lower'] = companies_df['permalink'].str.lower()
rounds2_df['company_permalink_lower'] = rounds2_df['company_permalink'].str.lower()

# merging the dataframes
master_frame = rounds2_df.merge(companies_df, how='outer', left_on='company_permalink_lower', right_on='permalink_lower')
master_frame.info()
```

# Table 2.1
#### The 2nd sheet in the excel file.

```python
# to find the value of various funding types, applying group by to the 'funding_round_type' field.

funding_type_df = master_frame.groupby('funding_round_type')

#----------------------------------------------------------------------------------------------------------------------

# getting the mean value for different types of funding. The apply() function is for showing the values in float format
# else the default view is in exponential format and hence unreadable.

funding_type_df['raised_amount_usd'].mean().sort_values(ascending=False).apply(lambda x: format(x, 'f'))
```

# Table 3.1

```python
# Finding the top 3 english speaking country.

# Grouping the companies_df dataframe by country_code, and then finding the sum of the raised_amount_usd in descending order.
english_country_df = master_frame.groupby('country_code')
english_country_df['raised_amount_usd'].sum().sort_values(ascending=False).apply(lambda x: format(x, 'f'))
```
